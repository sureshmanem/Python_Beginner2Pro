# MATPLOTLIB WITH OBJECT ORIENTED PROGRAMMING

- matplotlib creates a Figure instance, which is an object describing the plot window and its properties, and containing lists of all its elements
- matplotlib creates an Axes element within the figure. An axes can be thought of as a plotting window, where data can be arranged by x and y coordinates.
- add_axes(rectangle), where rectangle is a list [x0, y0, width, height] denoting the lower left point of the new axes in figure coodinates (x0,y0) and its width and height. 



```python
fig  = plt.figure()  # a new figure window
axes = fig.add_axes([0, 0, 1, 1])  # Add new axes

x = np.linspace(0, 10, 100)
y = np.sin(x)
axes.plot(x, y, 'r--', label = "Red Sine Wave #1")
axes.set_xlabel('Time')          # X-label 
axes.set_ylabel('Sine Wave')     # y-label
axes.set_title('My first Plot!') # Add a title
axes.legend()                    # Legend
axes.set_ylim([-1, 1])        # set x and y limits
axes.set_xlim([0, 10])    
axes.grid()
```

```python
# Creates a new figure
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # main axes
axes2 = fig.add_axes([0.2, 0.2, 0.5, 0.5]) # inset axes
axes3 = fig.add_axes([0.6, 0.2, 0.5, 0.5]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(x, y**2, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');

# Insert Figure Axes 3
axes3.plot(x, y**2, 'g')
axes3.set_xlabel('X_label_axes3')
axes3.set_ylabel('Y_label_axes3')
axes3.set_title('Axes 3 Title');
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1)
axes

# Use axes object to add elements to plot
axes[0].plot(x, y, 'r')
axes[0].set_xlabel('x1')
axes[0].set_ylabel('y1')
axes[0].set_title('title 1');

axes[1].plot(x, y**2, 'b')
axes[1].set_xlabel('x2')
axes[1].set_ylabel('y2')
axes[1].set_title('title 2');

axes[2].plot(x, y**3, 'b')
axes[2].set_xlabel('x3')
axes[2].set_ylabel('y3')
axes[2].set_title('title 3');
```

# ADJUST FIGURE SIZE, SPACING, COLOR

- plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.4)
    - left: left side of the subplots of the figure
    - right: right side of the subplots of the figure
    - bottom: bottom of the subplots of the figure
    - top: top of the subplots of the figure
    - wspace: amount of width reserved for space between subplots, expressed as a fraction of the average axis width
    - hspace: amount of height reserved for space between subplots, expressed as a fraction of the average axis height

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1, figsize=(10,10))
axes

# Use axes object to add elements to plot
axes[0].plot(x, y, 'r')
axes[0].set_xlabel('x1')
axes[0].set_ylabel('y1')
axes[0].set_title('title 1');

axes[1].plot(x, y**2, 'b')
axes[1].set_xlabel('x2')
axes[1].set_ylabel('y2')
axes[1].set_title('title 2');

axes[2].plot(x, y**3, 'b')
axes[2].set_xlabel('x3')
axes[2].set_ylabel('y3')
axes[2].set_title('title 3');


plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.4)
```

Reference: https://stackoverflow.com/questions/8409095/matplotlib-set-markers-for-individual-points-on-a-line

```python
fig, axes = plt.subplots(figsize=(15,15))

t = np.linspace(0, 10, 100)

# Lin width options
axes.plot(t, np.sin(t) + 1, color="blue", lw = 0.25)
axes.plot(t, np.sin(t) + 2, color="blue", lw = 1)
axes.plot(t, np.sin(t) + 3, color="blue", lw = 5)
axes.plot(t, np.sin(t) + 4, color="blue", lw = 9)

# Line Style options
axes.plot(t, np.sin(t) + 6, color="purple", lw = 5, ls ='-')
axes.plot(t, np.sin(t) + 7, color="purple", lw = 5, ls ='-.')
axes.plot(t, np.sin(t) + 8, color="purple", lw = 5, ls =':')


# Markers options
axes.plot(t, np.sin(t) + 10, color="red", lw = 0.1, ls='-', marker='s')
axes.plot(t, np.sin(t) + 11, color="red", lw = 0.1, ls='-', marker='o')
axes.plot(t, np.sin(t) + 12, color="red", lw = 0.1, ls='-', marker='*')
axes.plot(t, np.sin(t) + 13, color="red", lw = 0.1, ls='-', marker='+')


# marker size and color
axes.plot(t, np.sin(t) + 15, color="green", lw=1, ls='-', marker='o', markersize=3)
axes.plot(t, np.sin(t) + 16, color="green", lw=1, ls='-', marker='o', markersize=7)
axes.plot(t, np.sin(t) + 17, color="green", lw=1, ls='-', marker='s', markersize=10, markerfacecolor="yellow")
axes.plot(t, np.sin(t) + 18, color="green", lw=1, ls='-', marker='^', markersize=10, markerfacecolor="orange", markeredgewidth=2, markeredgecolor="blue");
```

```python
mu = 20 # mean of distribution
sigma = 3 # standard deviation of distribution
x = mu + sigma * np.random.randn(10000)
 
num_bins = 30

n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)

plt.ylabel('Probability')
plt.title(r'Histogram: $\mu=20$, $\sigma=3$')
 
```

```python
# numpy.random.normal() takes three arguments: mean, standard deviation of the normal distribution, and number of values desired.

np.random.seed(20)

data_1 = np.random.normal(200, 20, 2000)
data_2 = np.random.normal(60, 30, 2000)
data_3 = np.random.normal(70, 20, 2000)
data_4 = np.random.normal(40, 5, 2000)

data_all = [data_1, data_2, data_3, data_4]


fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)
bp = ax.boxplot(data_all)
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111, projection='3d')

# Grab some test data.
X, Y, Z = axes3d.get_test_data(0.05)

# Plot a basic wireframe.
ax.plot_wireframe(X, Y, Z, rstride=20, cstride=10)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure(figsize=(20,15))
ax = fig.gca(projection='3d')

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)

```
